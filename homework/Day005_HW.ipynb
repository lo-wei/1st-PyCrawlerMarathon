```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests
import json

pop_r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = pop_r.text
data = json.loads(response)

print(f'é€™å€‹APIä¸€æ¬¡æœƒå›å‚³ {len(data)} ç­†è³‡æ–™')
print('æ¯ä¸€ç­†è³‡æ–™åŒ…å«çš„æ¬„ä½:')
print(data[0].keys())
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for d in data:
    print('æ¨™é¡Œ:', d['title'])
    print('è²¼æ–‡æ™‚é–“:', d['createdAt'])
    print('ç•™è¨€äººæ•¸:', d['commentCount'])
    print('æŒ‰è®šäººæ•¸:', d['likeCount'])
    print('~')
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

#ç†±é–€
comment_count = 0
like_count = 0
total_count = 0
for d in data:
    comment_count += d['commentCount']
    like_count += d['likeCount']
    total_count += 1
    
print('ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€:', comment_count/total_count)
print('ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€:', like_count/total_count)

#éç†±é–€
non_pop_r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response = non_pop_r.text
data = json.loads(response)

comment_count = 0
like_count = 0
total_count = 0
for d in data:
    comment_count += d['commentCount']
    like_count += d['likeCount']
    total_count += 1
    
print('éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€:', comment_count/total_count)
print('éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€:', like_count/total_count)
```
