## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

## 1. 取出今天所有的新聞

```python
# 打開瀏覽器
from selenium import webdriver
from bs4 import BeautifulSoup
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

import time
from datetime import datetime
today_time = str(datetime.today().date()).replace("-", "/")
while True:
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight),2000;")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    news = soup.find(class_="part_list_2").find_all('h3')
    last_time = news[-1].find(class_="date").text[:10]
    if last_time != today_time:
        break

for d in news:
    if d.find(class_="date").text[:10] == today_time:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
now_hr = news[0].find(class_="date").text[11:13]
twohr_ago = int(now_hr) - 2

for d in news:
    if d.find(class_="date").text[11:13] >= str(twohr_ago)\
    and d.find(class_="date").text[:10] == today_time:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium.webdriver.support.ui import Select

day = datetime.now().day - 3

selectSite = Select(browser.find_element_by_id("selD"))
selectSite.select_by_value(str(day))
browser.find_element_by_id('button').click()

while True:
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight),2000;")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    news = soup.find(class_="part_list_2").find_all('h3')
    last_hr = news[-1].find(class_="date").text[11:13]
    if int(last_hr) < 15:
        break

for d in news:
    if 15 <= int(d.find(class_="date").text[11:13]) <= 17:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
